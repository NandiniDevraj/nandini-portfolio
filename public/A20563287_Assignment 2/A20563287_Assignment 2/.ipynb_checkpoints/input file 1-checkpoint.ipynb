To create this model, I made the following assumptions:

Shared Boundary: An edge exists between two community areas if they share a border of any length. Two areas that meet at a single point (a corner) are not considered adjacent.

```python
from pathlib import Path
import geopandas as gpd

DATA_DIR = Path(r"C:\Users\anuja\OneDrive\Desktop\SEM 3\OSNA- Anuja\Assignment 02\data")

geojsons = list(DATA_DIR.glob("*.geojson"))
zipshps  = list(DATA_DIR.glob("*.zip"))

assert geojsons or zipshps, f"No boundary file found in {DATA_DIR}"

# prefer the most recently modified geojson; otherwise the newest zip
if geojsons:
    IN_FILE = max(geojsons, key=lambda p: p.stat().st_mtime)
    gdf = gpd.read_file(IN_FILE)
else:
    IN_FILE = max(zipshps, key=lambda p: p.stat().st_mtime)
    gdf = gpd.read_file(f"zip://{IN_FILE}")

print("Using:", IN_FILE)
```

# compare degree distributions (Chicago vs class)

```python
import pandas as pd, networkx as nx
from pathlib import Path

ROOT = Path(r"C:\Users\anuja\OneDrive\Desktop\SEM 3\OSNA- Anuja\Assignment 02")
CHI  = ROOT/"out/chicago"
CLS  = ROOT/"out/class"

# Chicago graph
chi_edges = pd.read_csv(CHI/"1a_chicago_graph_input.csv")
G_chi = nx.from_pandas_edgelist(chi_edges, "area_u", "area_v")
chi_deg = [d for _,d in G_chi.degree()]
print("Chicago  nodes:", G_chi.number_of_nodes(), "edges:", G_chi.number_of_edges(),
      "avg degree:", round(sum(chi_deg)/len(chi_deg),2))

# Class unimodal graph (we didnt save its edgelist; just read degrees from the pictures graph)
# Rebuild quickly from 2a edges to be precise:
edges_df = pd.read_csv(CLS/"2a_bipartite_input.csv")
from collections import defaultdict
G = nx.Graph(); people = sorted(edges_df["person"].unique()); G.add_nodes_from(people)
by_ent = defaultdict(list)
for _,r in edges_df.iterrows(): by_ent[r["entity"]].append(r["person"])
for plist in by_ent.values():
    for i in range(len(plist)):
        for j in range(i+1, len(plist)):
            u,v = sorted((plist[i], plist[j]))
            G.add_edge(u,v, weight=G[u][v]["weight"]+1 if G.has_edge(u,v) else 1)

class_deg = [d for _,d in G.degree()]
print("Class  nodes:", G.number_of_nodes(), "edges:", G.number_of_edges(),
      "avg degree:", round(sum(class_deg)/len(class_deg),2))
```
